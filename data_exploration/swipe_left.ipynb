The swipe left gesture is done with the right arm, so we need to check, if there are differences in the movement between the body sides (left and right), to check for features, that are relevant for this gesture. We would expect the most movements to be in the x-value of the right hand and shoulder, indicating a movement from right to left.

# Summary

For the detection of the swipe left-gesture following body parts are relevant
- head, left shoulder, elbow (z): increase
- (elbow) hands (x, y, z): x is increasing, y is decreasing

# Statistics

```python
gesture_length = imp.get_gesture_length(Gesture.SWIPE_LEFT.value)
print(f"Shortest gesture: {min(gesture_length)} frames")
print(f"Longest gesture: {max(gesture_length)} frames")
print(f"Mean length: {round(np.mean(gesture_length), 2)} frames")
print(f"Median length: {np.median(gesture_length)} frames")
```
