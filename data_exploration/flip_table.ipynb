The flip table gesture is done with the both arms moving from bottom to top, so we need to check, if there are movements going upwards (in particular in hands and elbwos), to check for features, that are relevant for this gesture. As we already have seen in the up movement there could also be a small movement in the head-section going backwards (z) at the end of this gesture (when the arms don't have enough space to go further back and so you move your whole upper body backwards).

# Summary

For the detection of the flip table-gesture following body parts are relevant
- head & shoulders (z):  increase in the first frames
- elbows & hands (y, z): moving up (y) and first foreward, than backward (z)
