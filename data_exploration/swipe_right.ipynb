The swipe right gesture is done with the left arm, so we need to check, if there are differences in the movement between the body sides (left and right), to check for features, that are relevant for this gesture. We would expect the most movements to be in the x-value of the left hand and shoulder, indicating a movement from left to right.

# Summary

For the detection of the swipe right-gesture following body parts are relevant
- shoulder, elbows, hands for rotation of the body (z)
- left hand (x, y)

# Statistics

```python
gesture_length = imp.get_gesture_length(Gesture.SWIPE_RIGHT.value)
print(f"Shortest gesture: {min(gesture_length)} frames")
print(f"Longest gesture: {max(gesture_length)} frames")
print(f"Mean length: {round(np.mean(gesture_length), 2)} frames")
print(f"Median length: {np.median(gesture_length)} frames")
```
