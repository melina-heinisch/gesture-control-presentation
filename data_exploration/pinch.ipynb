The pinch gesture is done with both arms moving towards each other, so we would expect the gesture to show patterns on the x- and y-values of both arms.

# Summary

For the detection of the pinch-gesture following body parts are relevant
- head, shoulders (x,z): increasing
- left elbow (y): increasing
- elbow (z): decreasing
- hands (x, y, z)

## Statistics

```python
gesture_length = imp.get_gesture_length(Gesture.PINCH.value)
print(f"Shortest gesture: {min(gesture_length)} frames")
print(f"Longest gesture: {max(gesture_length)} frames")
print(f"Mean length: {round(np.mean(gesture_length), 2)} frames")
print(f"Median length: {np.median(gesture_length)} frames")
```
