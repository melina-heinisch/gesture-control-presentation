The spin gesture is done with both arms, continously rotating around each other, so we need to check, if there are any visible movements that could be associated with this gesture. We would expect movements in both arms that are visible in the forward/backward (z) and up/down (y) directions.

# Summary

For the detection of the spin-gesture following body parts are relevant
- Head & left shoulder, right elbow: wave movement, could be noise, but who knows? (z)
- hands: wave movements (y, z)
