The rotate left gesture is done with the left arm, so we need to check, if there are differences in the movement between the body sides (left and right), to check for features, that are relevant for this gesture. 

# Summary

For the detection of the rotate left-gesture following body parts are relevant
- Hill shape at the beginning: head & shoulder (z)
- small curves: left elbow & hand (x, y)
- decreasing / increasing: both elbows & hands (z)
