### Head

```python
all_frames_concat[["nose_z", 
                   "left_eye_inner_z", "left_eye_z", "left_eye_outer_z", 
                   "right_eye_inner_z", "right_eye_z", "right_eye_outer_z",
                   "left_mouth_z",  "right_mouth_z",                   
                   "left_ear_z", "right_ear_z", 
                   ]].corr()
```

As we can see here, all points of the head are highly correlating with each other. So it shouldn't be a problem to reduce those features to one feature.

```python
all_frames_concat[["nose_z", 
                   "left_eye_z", #the eyes are reduced to one point each, because the three points are completly redundant
                   "right_eye_z",
                   "left_mouth_z",  "right_mouth_z",                   
                   "left_ear_z", "right_ear_z", 
                   ]].corr().mean()
```

The highest correlation with all points of the face has the right mouth, which will now be used as a representant of the head-value.

### Other body parts

The elbow is still highly correlating (but not complete redundant with the hands). If it would be necessary to reduce the features even more, getting rid of the z-values of the elbows could be an option.

```python
all_frames_concat[["left_thumb_y", 
                   "left_elbow_y",
                   "right_thumb_y",
                   "right_elbow_y"
                   ]].corr()
```

```python
all_frames_concat[["right_mouth_z",
                   "left_shoulder_z",
                   "right_shoulder_z",
                   "left_thumb_z", 
                   "left_elbow_z",
                   "right_thumb_z",
                   "right_elbow_z"
                   ]].corr()
```

The shoulders are correlating to the head, but maybe there could be some additional information.
