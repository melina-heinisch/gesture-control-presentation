The up gesture is done with the right arm, so we need to check, if there are differences in the movement between the body sides (left and right), to check for features, that are relevant for this gesture. 

# Summary

For the detection of the up-gesture following body parts are relevant
- head & shoulder (z)
- right elbow (y & z)
- right hand (y & z)

# Statistics

```python
gesture_length = imp.get_gesture_length(Gesture.UP.value)
print(f"Shortest gesture: {min(gesture_length)} frames")
print(f"Longest gesture: {max(gesture_length)} frames")
print(f"Mean length: {round(np.mean(gesture_length), 2)} frames")
print(f"Median length: {np.median(gesture_length)} frames")
```
